```python
import numpy as np
import matplotlib.pyplot as plt
```

__Центральная предельная теорема__

1. Выбрать непрерывное распределение. 
2. Сгенерировать из него выборку объёма 1000, построить гистограмму выборки и нарисовать поверх неё теоретическую плотность распределения вашей случайной величины.
3. Оценить распределение выборочного среднего вашей случайной величины при разных объёмах выборок. Для этого при трёх и более значениях n (например, 5, 10, 50) сгенерировать 1000 выборок объёма n и построить гистограммы распределений их выборочных средних. 
4. Используя информацию о среднем и дисперсии исходного распределения, посчитать значения параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних. Обратить внимание: для подсчёта значений этих параметров нужно использовать именно теоретические среднее и дисперсию вашей случайной величины, а не их выборочные оценки.
5. Описать разницу между полученными распределениями при различных значениях n. Как меняется точность аппроксимации распределения выборочных средних нормальным с ростом n?

```python
# 1
from scipy.stats import uniform, norm

a = 3
b = 10
u = uniform(loc=a, scale=b-a)  # X ~ U(3, 10)
```

```python
# 2
sample = u.rvs(size=1000)

plt.figure(figsize=(12, 6))
plt.hist(sample, bins=20, label="Гистограмма", density=True)
x = np.linspace(0,12,100)
plt.plot(x, u.pdf(x), linewidth=3, label='Теоретическая плотность распределения')
plt.legend()
plt.title("Сравнение экспериментальной и теоретической функции распределения")
plt.show()
```

```python
# 3
n_dict = {
    5: [],
    10: [],
    50: []
}
for n in n_dict:
    for _ in range(1000):
        current_sample = u.rvs(size=n)
        mean = current_sample.mean()
        n_dict[n].append(mean)
```

Согласно ЦПТ $\bar X \sim N(EX, \dfrac{DX}{n})$

```python
# 4
EX = (a + b)/2
DX = (b - a) ** 2 / 12
```

```python
# 3, 4
n = 5
plt.figure(figsize=(12, 6))
sample = n_dict[n]
plt.hist(sample, bins=20, label="Гистограмма для 1000 выборок", density=True)
x = np.linspace(3,10,100)
gauss = norm(loc=EX, scale=np.sqrt(DX/n))
plt.plot(x, gauss.pdf(x), linewidth=3, label='Плотность распределения по ЦПТ')
plt.title(f"Гистограмма для n = {n}")
plt.legend()
plt.show()
```

```python
# 3, 4
n = 10
plt.figure(figsize=(12, 6))
sample = n_dict[n]
plt.hist(sample, bins=20, label="Гистограмма для 1000 выборок", density=True)
x = np.linspace(3,10,100)
gauss = norm(loc=EX, scale=np.sqrt(DX/n))
plt.plot(x, gauss.pdf(x), linewidth=3, label='Плотность распределения по ЦПТ')
plt.title(f"Гистограмма для n = {n}")
plt.legend()
plt.show()
```

```python
# 3, 4
n = 50
plt.figure(figsize=(12, 6))
sample = n_dict[n]
plt.hist(sample, bins=20, label="Гистограмма для 1000 выборок", density=True)
x = np.linspace(3,10,100)
gauss = norm(loc=EX, scale=np.sqrt(DX/n))
plt.plot(x, gauss.pdf(x), linewidth=3, label='Плотность распределения по ЦПТ')
plt.title(f"Гистограмма для n = {n}")
plt.legend()
plt.show()
```

## Выводы
С увеличением n растет точность апроксимации и гистограмма всё больше начинает совпадать с графиком функции плотности нормального распределения по ЦПТ
